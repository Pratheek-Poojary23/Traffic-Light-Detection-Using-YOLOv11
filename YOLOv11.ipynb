```python
import ultralytics
from ultralytics import YOLO
```

```python
import torch
print("CUDA Available:", torch.cuda.is_available())
print("GPU:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")
```

1 Training

```python
model= YOLO('yolo11n.pt')
model.train(data= 'dataset_custom.yaml',imgsz= 640, 
            batch=8, epochs=100, device=0, workers=0)
```

2.Prediction

```python
model= YOLO('YOLOv11_custom.pt')
model.predict(source="2.jpg", show=True, save= True, conf= 0.5,
              show_labels=True, show_conf=True,classes=[0,1])
```
